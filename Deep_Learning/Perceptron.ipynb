### Independent and Dependent Variable Splitting

```python
placement.shape
```

```python
placement.head()
```

```python
X = placement.iloc[:,0:2].values
print(type(X))
Y = placement.iloc[:,-1].values
print(type(Y))
```

### get the perceptron model 

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
```

```python
p.fit(X,Y)
```

### getting the w1 and w2 values using perceptron coefficient

```python
p.coef_
```

```python
p.intercept_
```
